```python
#http://nbviewer.jupyter.org/github/python-visualization/folium/blob/master/examples/GeoJSON_and_choropleth.ipynb?flush_cache=true


colormap = linear.OrRd.scale(0.0,1.0)
colormap
```

```python
sfmap = folium.Map([37.7556, -122.4399], zoom_start = 12)
folium.GeoJson(gdf1.to_json(),overlay=True,
        style_function=lambda feature: 
               {'color': "black",
               'weight':1.5,
               'fillColor': colormap(gdf[feature['id']])}
              ).add_to(sfmap)
folium.LayerControl().add_to(sfmap)
```

```python
def add_ambmarker(lat,long,amb_name):
    folium.Marker([lat, long], icon=folium.Icon(icon='plus',color='blue'),
                  popup="Amb #:" + amb_name
                 ).add_to(sfmap)
```

```python
def add_emsmarker(lat,long,event_id):
    folium.RegularPolygonMarker([lat, long], popup="EMS #:" + event_id,
                                fill_color='red',number_of_sides=5,radius=10).add_to(sfmap)
```

```python
def dispatch_ambmarker(lat,long,amb_name):
    folium.Marker([lat, long], icon=folium.Icon(icon='plus',color='grey'),
                  popup="Amb #:" + amb_name
                 ).add_to(sfmap)
```

```python
add_ambmarker(37.7556, -122.4399,'13')
add_ambmarker(37.7089, -122.4404,'24')
add_ambmarker(37.8016, -122.4182,'35')
add_ambmarker(37.7862, -122.4198,'46')
add_emsmarker(37.7862, -122.4298,'123')
```
